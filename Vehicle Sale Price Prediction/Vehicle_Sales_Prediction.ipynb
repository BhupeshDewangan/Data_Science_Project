```python
car = car.reset_index(drop = True)
```

```python
x = car[['name', 'year', 'km_driven', 'fuel']]
y = car['selling_price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name', 'fuel']])
```

```python
OneHotEncoder()
```

```python
make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name', 'fuel']), remainder='passthrough')
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name', 'fuel']), remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
scores = []
for i in range(850):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred)
```
